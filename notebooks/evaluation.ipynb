```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ticks = [50, 100, 150, 200, 250, 300]
losses = {"hinge": [], "sigmoid": []}

for loss in losses.keys():
    for n in ticks:
        print(f"{loss}_{n}_alignment_logits")
        df = pd.read_csv(f'../results/{loss}_{n}_alignment_logits.csv', converters={'logits': pd.eval})
        logits = df.logits.values
        mean_logit = [sum([x[0] for x in logits]) / len(logits), sum([x[1] for x in logits]) / len(logits)]
        losses[loss].append(mean_logit)
```

```python
print("no_alignment_logits")
df = pd.read_csv(f'../results/no_alignment_logits.csv', converters={'logits': pd.eval})
logits = df.logits.values
no_alignment_logits = [sum([x[0] for x in logits]) / len(logits), sum([x[1] for x in logits]) / len(logits)]
no_alignment_logits
```
